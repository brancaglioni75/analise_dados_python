```python
# Quantidade de Aptos a votar
Qtd_Aptos_Votar = df['QT_APTOS_TOT'].sum()
print(Qtd_Aptos_Votar)
```

```python
# Quantidade de Votos de Lula
Qtd_Votos_lula = df['LULA_QT_VOTOS_TOT'].sum()
print(Qtd_Votos_lula)
```

```python
# Quantidade de Votos de Bolsonaro
Qtd_Votos_Bolsonaro = df['JAIR_BOLSONARO_QT_VOTOS_TOT'].sum()
print(Qtd_Votos_Bolsonaro)
```

```python
# % de Votos de Lula
Percentual_Votos_Lula = (Qtd_Votos_lula/(Qtd_Votos_Bolsonaro+Qtd_Votos_lula)*100).round(2)
print(Percentual_Votos_Lula)
```

```python
# % de Votos de Bolsonaro
Percentual_Votos_Bolsonaro = (Qtd_Votos_Bolsonaro/(Qtd_Votos_Bolsonaro+Qtd_Votos_lula)*100).round(2)
print(Percentual_Votos_Bolsonaro)
```

```python
# Quantidade de Votos em Branco
Qtd_Votos_Branco = df['BRANCO_QT_VOTOS_TOT'].sum()
print(Qtd_Votos_Branco)
```

```python
# Quantidade de Votos Nulos
Qtd_Votos_Nulos = df['NULO_QT_VOTOS_TOT'].sum()
print(Qtd_Votos_Nulos)
```
